```python
# sum of list using function
def s_function(x):
    sm=0
    for t in x:
        sm=sm+t
    return sm

a=[23,34,23,4,23]
sm=s_function(a)
print(sm)
        
print('---')

# finding absolute value using function
def s_abs(x):
    if x<0:
        return -x
    else: return x

# n=-80
# abs=s_abs(n)
# print(abs)
print(s_abs(-80))

print('---')

#finding factorial using function(using range)
def s_fac(x):
    fact=1
    for t in range(1,x+1):
        fact=t*fact
    return fact
print(s_fac(5))

#factorial using while loop
def s_fact(x):
    fac=1
    while x>0:
        fac=fac*x
        x=x-1
    return fac
print(s_fact(5))

print('---')
#check whether the number is palandrome or not
def s_pal(x):
    ans=0
    temp=x
    while x>0:
        rem=x%10
        ans=10*ans+rem
        x=x//10
    if temp==ans:
        return True
    return False
print(s_pal(545))
print(s_pal(245))
print(s_pal(0))      

#check whether the number is palandrome or not using string
def s_pala(x):
    if int(str(x)[::-1])==x:
        return True
    return False

print(s_pala(252))

print('---')

#check whether a string is palandrome or not
ans="bob"
if ans==ans[::-1]:
    print("palandrome")
else: print("not palandrome")
```
